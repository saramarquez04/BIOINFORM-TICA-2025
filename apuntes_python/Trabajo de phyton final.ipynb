#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
import re
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
```

```python
#para realizar este paso he utilizado las expresiones regulares
#1º aquí estoy creando una función con el código def llamada digestar_secuencia con los parámetros (secuencia,enzimas):, secuencia es mi cadena de ADN que quiero analizar y enzimas es un diccionario con las enzimas del enunciado y sus sitios corte
# este código nos sirve para cuando más adelante pongamos la función "resultado = digestar_secuencia(secuencia_adn, enzR)" phyton entienda secuencia es secuencia_adn y enzimas enzR
#después creamos 3 diccionarios en los que se guardarán las enzimas con sitio de corte, las enzimas sin sitio de corte y posiciones de corte, estos diccionarios están formados por una key y un valor, la key es el nombre de nuestra enzima y el valor el punto de mi secuencia de adn en el que cambia
# es decir, como más adelante se verá "HindIII: [3373, 7619]" esto significa que nuestra enzima HindIII corta en la posición 3373 y 7619 de la secuencia

def digestar_secuencia(secuencia, enzimas):
#dicionario para guardar enzimas que si cortan en algun punto de nuestra secuencia de adn 
    enzimas_con_corte = {}
#lista para almacenar enzimas que no cortan
    enzimas_sin_corte = []
#lista para guardar todas las posiciones de corte de todas las enzimas propuestas
    posiciones_de_corte = []

#al poner el código "enzimas.items() estamos creando parejas (como una lista de tuplas) dentro del diccionario (key y valor, o en este caso enzima y siito de corte) 
#por lo tanto, con este bucle creado recorremos todos los valores de ese diccionario y desempaquetando las tuplas en las dos variables que ya he nombrado
#una vez que tenemos la dos variables el comando "posicion_corte = secuencia_corte.index('*') y "secuencia_real = secuencia_corte.replace('*', '')" nos sirven para encontrar el asterisco (que es la posición de corte de la enzima) y quitar el asterisco para quedarse con la secuencia real que se debe buscar en nuestra secuencia_adn respectivamente
    for enzima, secuencia_corte in enzimas.items():
        posicion_corte = secuencia_corte.index('*')
        secuencia_real = secuencia_corte.replace('*', '')

#con estos comandos estamos buscando los sitios donde la enzima puede cortar, si encuentra alguno se guardan las posiciones de corte y si no, se marca que la enzima no corta
#con re.finditer estamos buscando todas las veces que aparece la secuencia que reconoce la enzima dentro de nuestra secuencia de adn
#con el código "match.start" obtenemos el sitio de inicio donde la secuencia de la enzima ha sido encontrada y con "posición_corte" obtenemos el desplazamiento del corte con respecto al inicio del sitio de corte 
#por lo tanto al combinar "match.start" y "posicion_corte" obtenemos la posición precisa del corte en la secuencia del adn      
        
        
#con el código if cortes: se clasifica y recogen los puntos de corte en la lista posiciones_de_corte 
        matches = re.finditer(secuencia_real, secuencia)
        cortes = [match.start() + posicion_corte for match in matches]
        if cortes:
            enzimas_con_corte[enzima] = cortes
            posiciones_de_corte.extend(cortes)
        else:
            enzimas_sin_corte.append(enzima)

#con la función sort estamos ordenando los puntos de corte de manera ascendente lo cual sirve para ordenar dentro de la lista según el punto de corte que aparece 1º y los que van apareciendo después
#con el bucle "for" obtenemos los puntos de corte intermedios y con "fragmentos.append(secuencia[inicio:corte])" obtenemos el último
#con "else" obtenemos los casos en los que no hay puntos de corte    
    if posiciones_de_corte:
        posiciones_de_corte.sort() 
        fragmentos = []
        inicio = 0
        
        for corte in posiciones_de_corte:
            fragmentos.append(secuencia[inicio:corte])
            inicio = corte
        
        fragmentos.append(secuencia[inicio:])
    else:
        fragmentos = [secuencia]

#para obtener los resultados en la carpeta de salida utilizamos un diccionario para extraer la información (enzimas con corte, enzimas sin corte, sus fragmentos y sus tamaños)
    tamaños_fragmentos = [len(fragmento) for fragmento in fragmentos]

    return {
        "enzimas_con_corte": enzimas_con_corte,
        "enzimas_sin_corte": enzimas_sin_corte,
        "fragmentos_resultantes": fragmentos,
        "tamaños_fragmentos": tamaños_fragmentos
    }

resultado = digestar_secuencia(secuencia_adn, enzR)
print("\033[1mEnzimas que tienen sitios de corte en la secuencia y posiciones:\033[0m")
for enzima, posiciones in resultado["enzimas_con_corte"].items():
    print(f"{enzima}: {posiciones}")
print("\033[1mEnzimas que NO tienen sitios de corte:\033[0m")
print(resultado["enzimas_sin_corte"])
print("\033[1mFragmentos resultantes de la digestión:\033[0m")
for fragmento, tamano in zip(resultado["fragmentos_resultantes"], resultado["tamaños_fragmentos"]):
    print(f"FRAGMENTO: {fragmento}, TAMAÑO: {tamano}")
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
#para encontrar los ORFs y calcular el % de aminoácidos aromáticos es necesario traducir nuestra secuencia de ADN para que mediante el proceso de traducción la hebra de ARNm se convierta en proteina leyendo los codones (secuencias de tripletes)
#para ellos he creado un diccionario llamado "tabla_codones" el cual cuenta con una key (que muestra el triplete de ADN) y con un valor (que es el aminoácido que codifica ese triplete como metionina o * para codón de parada), esto quiere decir que si al ejecutar el código se encuentra el triplete ATG pues se obtendría el valor metionina y por tanto se podría traducir la cadena

tabla_codones = {
    'ATA':'I','ATC':'I','ATT':'I','ATG':'M',
    'ACA':'T','ACC':'T','ACG':'T','ACT':'T',
    'AAC':'N','AAT':'N','AAA':'K','AAG':'K',
    'AGC':'S','AGT':'S','AGA':'R','AGG':'R',
    'CTA':'L','CTC':'L','CTG':'L','CTT':'L',
    'CCA':'P','CCC':'P','CCG':'P','CCT':'P',
    'CAC':'H','CAT':'H','CAA':'Q','CAG':'Q',
    'CGA':'R','CGC':'R','CGG':'R','CGT':'R',
    'GTA':'V','GTC':'V','GTG':'V','GTT':'V',
    'GCA':'A','GCC':'A','GCG':'A','GCT':'A',
    'GAC':'D','GAT':'D','GAA':'E','GAG':'E',
    'GGA':'G','GGC':'G','GGG':'G','GGT':'G',
    'TCA':'S','TCC':'S','TCG':'S','TCT':'S',
    'TTC':'F','TTT':'F','TTA':'L','TTG':'L',
    'TAC':'Y','TAT':'Y','TAA':'','TAG':'',
    'TGC':'C','TGT':'C','TGA':'*','TGG':'W',
}
```

```python
#con el código "deftraducir(dna)" estoy definiendo la función
#con la función "proteina="" se inicia la cadena que irá almacenando la secuencia de aminoácidos
#como he mencionado hemos agrupado los aminoácidos en grupos de 3 con el objetivo de luego traducirlo a proteinas y para esto usamos la función "for i in range(0, len(dna)-2, 3):" para iniciar un bucle que recorra la secuencia de adn de 3 en 3 para generar esos codones
#para simular el proceso de traducción extraemos un codón con la función "codon = dna[i:i+3]" y el programa consulta nuestra tabla de codones para obtener el aminoácido correspondiente mediante el comando "aa = tabla_codones.get(codon, "X")"
#por último en este paso se agrega el aminoácido o codón de parada (*) correspondiente a nuestra cadena que se ha llamado como proteína con la función "proteina+=aa" y se devuelve esta secuencia traducida a proteina con "return proteina"

def traducir(dna):
    proteina = ""
    for i in range(0, len(dna)-2, 3):
        codon = dna[i:i+3]
        aa = tabla_codones.get(codon, "X")
        proteina += aa
    return proteina

#con estos códigos estamos generando la cadena complementaria inversa
#para ello, igual que he hecho antes 1º se define la función con el comando "def" y creamos un diccionario para refeljar los cambios de aminoácidos que se ven al crear dicha cadena complementaria (es decir, cambio de a - T; T-a; G-C y G-C"
#con la función return unimos los caracteres, con la función ".join(complemento.get(base, "N")" para cada base buscamos la complementaria o añadimos N en el caso de que dicha base no esté entre los caracteres que he nombrado arriba, y con "for base in reversed(seq))" basicamente le estamos diciendo al programa que itere sobre la secuencia de ADN desde el final hacia el principio (ya que es la cadena complementaria inversa)

def complementaria_inversa(seq):
    complemento = {'A':'T','T':'A','C':'G','G':'C'}
    return "".join(complemento.get(base, "N") for base in reversed(seq))

#volvemos a definir la función y creamos una lista que almacenará los ORFS que se encuentren:
def analizar_orfs(secuencia):
    orfs = []

#creamos un bucle para los 3 marcos de lectura con "for i in range(3):" y con "prot = traducir(secuencia[i:])" traducimos el ADN
#para dividir la lista de la proteina en una lista de fragmentos utilizo el codón de parada como separador con "for orf in prot.split("*"):"
#por último le decimos al programa que solo almacene aquellos fragmentos que tengan una longuitud mayor a 0 y que guarde su longuitud y hebra 
    for i in range(3):
        prot = traducir(secuencia[i:])
        for orf in prot.split("*"):
            if len(orf) > 0:
                orfs.append((orf, "+", len(orf)))

#calculamos la hebra complementaria reversa y repetimos lo explicado anteriormente para los 3 marcos de lectura existentes usando comp.
    comp = complementaria_inversa(secuencia)
    for i in range(3):
        prot = traducir(comp[i:])
        for orf in prot.split("*"):
            if len(orf) > 0:
                orfs.append((orf, "-", len(orf)))
                
    #ordenamos la lista de ORFS en base a su longuitud de mayor a menor
    orfs = sorted(orfs, key=lambda x: x[2], reverse=True)

#obtenemos la información del ORF más largo y contamos los aminoácidos aromáticos presentes por último calculamos la longitud relativa de esta ORF 

    mejor_orf, mejor_strand, mejor_len = orfs[0]

    aromaticos = mejor_orf.count("F") + mejor_orf.count("W") + mejor_orf.count("Y")
    porcentaje_aromaticos = (aromaticos / len(mejor_orf)) * 100 if mejor_len > 0 else 0
    longitud_relativa = (mejor_len * 3 / len(secuencia)) * 100

    resultados = {
        "\033[1mLos ORFs de la secuencia son\033[0m": len(orfs),
        "\033[1mEl mejor ORF es\033[0m": (mejor_orf, mejor_strand, mejor_len),
        "\033[1mEl porcentaje de aminoácidos aromáticos es\033[0m": round(porcentaje_aromaticos, 2),
        "\033[1mY su longitud relativa (%)\033[0m": round(longitud_relativa, 2)
    }
#obtenemos el diccionario de resultados y la lista completa de ORFS ya que queremos imprimir los 6 más largos
    return resultados, orfs

#con este código importamos textwrap que es una librería necesaria para ejecutar la función posterior ya qe nos permite dividir las cadenas más largas en cadenas más cortas 
import textwrap

def imprimir_orfs_fasta(orfs, top=6, wrap=60):
    """
    Imprime los 'top' ORFs más largos.
    """
    top_orfs = sorted(orfs, key=lambda x: x[2], reverse=True)[:top]
    print(f"Los {top} ORFs más largos:")
    for idx, (seq, strand, length) in enumerate(top_orfs, start=1):
        print(f">ORF_{idx}|hebra={strand}|aa={length}|nt={length*3}")
        print("\n".join(textwrap.wrap(seq, width=wrap)))
        print()

resultado, orfs = analizar_orfs(secuencia_adn)
#imprimimos los resultados
for clave, valor in resultado.items():
    print(f"{clave}: {valor}")

imprimir_orfs_fasta(orfs, top=6, wrap=60)
```
