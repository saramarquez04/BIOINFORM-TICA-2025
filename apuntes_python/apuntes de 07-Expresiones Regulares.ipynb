# EXPRESIONES REGULARES

El módulo re contiene los métodos necesarios para analizar y buscar expresiones regulares en Python.

**import re**

**re.search(pattern,string)**

La función usa dos argumentos, el patrón y la cadena dónde hay que buscarlo


Para indicar a Python que ignore los caracteres especiales como *\t* o *\n* se puede usar *r* fuera de las comillas: **print(r”\t\n”)**

El resultado que devuelve **re.search()** es booleano, true/false

```python
#las expresiones regulares se usan con un módulo que es el re entonces siempre 1º tenemos que importar el módulo
#la idea de las expresiones regulares esque vamos a buscar un patrón de carácteres es una cadena 
#OJO, en el trabajo siempre hay que poner esto !!
import re
```

```python
# re.search(pattern,string)
# r se usa para que Python ignore caracteres especiales

#siempre ponemos 1º re.search (nuestro patrón, cadena donde vamos a buscarlo)
#lo que devuelve una expresión regular es un resultado de verdadero o falso es decir, o lo encuentra o no lo encuentra
```

```python
# aquí lo que queremos hacer es buscar el patrón GAZ dentro de la cadena de ADN
#ponemos re.search, luego lo que queremos buscar (el patrón) y luego la cadena donde lo queremos buscar, en este caso hemos llamado a la cadena dna asique ponemos dna
#la r es un recurso que se usa para ignorar caracteres especiales, recomendado ponerlo
#el resultado lo sacamos poniendo print en función de si lo encuentra o no lo encuentra, como no lo encuentra pues sale not site found

dna = "ATCGCGAATTGAFCAC"
if re.search(r"GAZ", dna):
    print("restriction enzyme site found!")
else:
    print("Not site found!")
```

```python
# Alternaciones
#cuando quieres buscar varias cosas (más de un patrón) pones el "o" con el "or"! o variaciones del mismo patrón
dna = "ATCGCGAATTCAC"
if re.search(r"ATT", dna) or re.search(r"CGAA", dna):
    print("restriction enzyme site found!")
```

```python
#la barra vertical es lo mismo que poner "o", entonces le está diciendo que en esa posición puede ir una G o una C, lo demás va seguro
#para poner el "o" siempre entre paréntesis y con la barra vertical!
dna = "ATCGCGAATTCAC"
if re.search(r"GC(G|C)AATT", dna):
    print("restriction enzyme site found!")
```

```python
# Grupos de caracteres
#lo mismo cuando quieres añadir muchas variaciones
dna = "ATCGCGGCAATTCAC"
if re.search(r"GC(A|T|G|C)GC", dna):
    print("restriction enzyme site found!")
```

```python
#también se pueden poner las alternativas todas juntas entre corchetes
dna = "ATCGCGGCAATTCAC"
if re.search(r"GC[ATGC]GC", dna):
    print("restriction enzyme site found!")	
```

```python
# Negaciones
#quieres añadir un patrón negativo, es decir, que empiece por GC pero que luego tenga cualquier cosa menos una N
# se pone con el sombrerito N = todo menos la N
dna = "ATCGCGAATTCAC"
if re.search(r"GC[^N]", dna):
    print("restriction enzyme site found!")
```

```python
# Extración de parte del patrón en una variable
dna = "ATGACGTACGTACGACTG"
# guarda los patrones en la variable m
m = re.search(r"GA[ATGC]{3}AC",dna)
#el 3 con la llave es una llave de repetición, es decir, uno de esos 3 veces
m_pattern = m.group()
print(m_pattern)

#\d (esto significa número) + [5] (cinco) = 5 números; . significa cualquier cosa, \. es cualquier punto literal, d+ es que el número este una vez como mínimo
#poner la barrita delante
#OJO ESTO SIRVE PARA EL TRABAJO CUANDO NOS PREGUNTA POR ENZIMAS DE RESTRICCIÓN!
#para guardar los patrones lo guardas en una variable de resultado que se llama m ( por ejemplo)
#la variable m engloba todos los patrones que has encontrado, la función .group las muestra?
```

```python
# Extrae un más de un bit del patrón?
dna = "ATGACGTACGTACGACTG"
m = re.search(r"GA([ATGC]{3})AC([ATGC]{2})AC", dna)
print("entire match: " + m.group())
print("first bit: " + m.group(1))
print("second bit: " + m.group(2))
```

```python
# Algunos métodos
dna = "ATGACGTACGTACGACTG"
m = re.search(r"GA([ATGC]{3})AC([ATGC]{2})AC", dna)
print("input string: " + dna)
print("entire match: " + m.group())
```

```python
# Start()
#nos dice en que posición comienza el patrón, importante en enzimas de restricción
print("start: " + str(m.start()))
```

```python
# End()
#nos dice donde acaba el patrón
print("end: " + str(m.end()))
```

```python
# Split()
#para dividir la cadena de entrada según el carácter que nosotros digamos, en este caso la C
dna = "ACTNGCATRGCTACGTYACGATSCGAWTCG"
runs = re.split(r"C", dna)
print(runs)	
```

```python
# Extraer  usando group()
dna = "ATGACGTACGTACGACTG"
m = re.search(r"GA([ATGC]{3})AC([ATGC]{2})AC", dna)
print("input string: " + dna)
print("entire match: " + m.group())
print("first bit: " + m.group(1))
print("second bit: " + m.group(2))
```

```python
# Finditer() detecta las bases erróneas 
#en este caso le estás diciendo que te identifique todas las cosas que no sean A,T,G,C
#TAMBIÉN NOS INTERESA PARA ENZIMAS DE RESTRICCIÓN!
dna = "CGCTCNTAGATGCGCRATGACTGCAYTGC" 

matches = re.finditer(r"[^ATGC]", dna) 
for m in matches: 
    base = m.group() 
    pos  = m.start() 
    print(base + " found at position " + str(pos))
```

```python
dna = "CGCTCNTAGATGCGCRATGACTGCAYTGC" 

matches = re.finditer(r"[^ATGC]", dna) 

for i in matches:
    match = i.group()
    pos = i.start()
    print(f"La base {match} está en la posición {pos}")
```

```python
# Findall()
#útil cuando hay patrones que se repiten varias veces, util para sacar las zonas de corte de una enzima de restricción y luego sacar las posiciones
#estos es una POLIA!!! los patrones de AT repetido 6 veces
dna = "CTGCATTATATCGTACGAAATTATACGCGCG" 
result = re.findall(r"[AT]{6,}", dna) 
print(result)
```
