```python
def analizar_A(secuencia=None, decimales=2):
    """
    Función principal que analiza el porcentaje de 'A' en una secuencia y usa una función interna encapsulada.
    
    Argumentos:
    secuencia : str : secuencia de ADN/ARN (por defecto una secuencia de ejemplo)
    decimales : int : número de decimales para el resultado (por defecto 2)
    
    Retorna:
    float : porcentaje de 'A' redondeado a los decimales indicados
    """

    # Secuencia por defecto si no se pasa ninguna
    if secuencia is None:
        secuencia = "ATCGATCGATCGATCGACTGACTAGTCATAGCTATGCATGTAGCTACTCGATCGATCGATCGATCGATCGATCGATCGATCGATCATGCTATCATCGATCGATATCGATGCATCGACTACTAT"
    
    # Función interna encapsulada
    def porcentaje_A(seq, dec=2):
        seq = seq.upper()  # manejar mayúsculas/minúsculas
        total = len(seq)
        if total == 0:
            return 0
        num_A = seq.count('A')
        return round((num_A / total) * 100, dec)
    
    # Llamamos a la función interna
    return porcentaje_A(secuencia, decimales)

# Ejemplos de uso
print("Sin argumentos:", analizar_A(), "%")              # usa secuencia y decimales por defecto
print("Con secuencia propia:", analizar_A("atgcatgca"), "%")  # decimales por defecto
print("Con secuencia y 4 decimales:", analizar_A("atgcatgca", 4), "%")
```

#### 5.- De la lista de accessions que se proporciona obtén lo siguiente:
1. Los que tengan el número 5
2. Los que tengan la letra 'd' o la 'e'
3. Los que tengan las letras 'd' y 'e' en ese orden
4. Los que tengan las letras 'd' y 'e' en ese orden con una letra cualquiera en medio
5. Los que tengan ambas letras 'd' y 'e' en cualquier orden
6. Los que empiezan por 'x' o por 'y'
7. Los que empiezan por 'x' o 'y' y terminan por 'e'
8. Los que tienen tres o más dígitos seguidos
9. Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'

```python
import re
```

```python
accs = ['xkn59438', 'yhdck2', 'eihd39d9', 'chdsye847', 'hedle3455', 'xjhd53e', '45da', 'de37dp']
# 1. Los que tengan el número 5
con_5 = [acc for acc in accs if '5' in acc]

# 2. Los que tengan la letra 'd' o la 'e'
con_d_o_e = [acc for acc in accs if 'd' in acc or 'e' in acc]

# 3. Los que tengan las letras 'd' y 'e' en ese orden
con_de = [acc for acc in accs if 'de' in acc]

# 4. Los que tengan 'd' y 'e' en ese orden con una letra cualquiera en medio
con_d_letra_e = [acc for acc in accs if re.search(r'd.e', acc)]

# 5. Los que tengan ambas letras 'd' y 'e' en cualquier orden
con_d_y_e = [acc for acc in accs if 'd' in acc and 'e' in acc]

# 6. Los que empiezan por 'x' o por 'y'
empieza_x_o_y = [acc for acc in accs if acc.startswith(('x', 'y'))]

# 7. Los que empiezan por 'x' o por 'y' y terminan por 'e'
empieza_x_o_y_termina_e = [acc for acc in accs if acc.startswith(('x', 'y')) and acc.endswith('e')]

# 8. Los que tienen tres o más dígitos seguidos
tres_digitos = [acc for acc in accs if re.search(r'\d{3,}', acc)]

# 9. Los que terminan con 'd' seguido de 'a' o 'r' o 'p'
termina_d_ar_p = [acc for acc in accs if re.search(r'd[arp]$', acc)]

# Mostrar resultados
print("1. Con número 5:", con_5)
print("2. Con 'd' o 'e':", con_d_o_e)
print("3. Con 'd' y 'e' en ese orden:", con_de)
print("4. 'd' y 'e' con una letra en medio:", con_d_letra_e)
print("5. Con 'd' y 'e' en cualquier orden:", con_d_y_e)
print("6. Empieza por 'x' o 'y':", empieza_x_o_y)
print("7. Empieza por 'x' o 'y' y termina por 'e':", empieza_x_o_y_termina_e)
print("8. Tres o más dígitos seguidos:", tres_digitos)
print("9. Termina con 'd' seguido de 'a', 'r' o 'p':", termina_d_ar_p)
```
